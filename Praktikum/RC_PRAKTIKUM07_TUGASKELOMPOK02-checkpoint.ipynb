```python
import csv
from functools import reduce

with open("C:/Users/rafif/Documents/cool yeah/SEMESTER 4/pbf/911_Calls_for_Service_(Last_30_Days).csv",'r') as file:
    data = list(csv.DictReader(file))
```

### Bagian 1: Model Populasi Polisi Detroit
Baca data dari file Laporan Polisi Detroit menggunakan modul csv dan terjemahkan data ke dalam daftar dictionary. Menggunakan Filter dengan fungsi lambda untuk mengecualikan dictionary (baris CSV) yang memiliki data yang hilang di kolom Zip, atau kolom Neighborhood. Dengan menggunakan fungsi lambda dan Reduce, hitung total waktu respons rata-rata, waktu pengiriman rata-rata, dan total waktu rata-rata untuk kepolisian Detroit.

```python
# Filter data yang tidak memiliki data yang hilang di kolom Zip atau Neighborhood
data_filtered = list(filter(lambda x: x['zip_code'] and x['neighborhood'] and x['totalresponsetime'] and x['dispatchtime'] and x['totaltime'], data))

# Menghitung total waktu respons rata-rata
total_response_time = reduce(lambda x, y: float(x) + float(y), [x['totalresponsetime'] for x in data_filtered])
average_response_time = total_response_time / len(data_filtered)

# Menghitung waktu pengiriman rata-rata
total_dispatch_time = reduce(lambda x, y: float(x) + float(y), [x['dispatchtime'] for x in data_filtered])
average_dispatch_time = total_dispatch_time / len(data_filtered)

# Menghitung waktu rata-rata kepolisian detroit
total_time = reduce(lambda x, y: float(x) + float(y), [x['totaltime'] for x in data_filtered])
average_time = total_time / len(data_filtered)


print("waktu respons rata-rata:", average_response_time)
print("waktu pengiriman rata-rata:", average_dispatch_time)
print("waktu rata-rata kepolisian detroit:", average_time )
```

### Bagian 2: Modelkan Neighborhood Samples
Menggunakan fungsi lambda dan map, atau lambda dan Filter, bagi list dictionary menjadi list dictionary yang lebih kecil yang dipisahkan oleh neighborhood. Dengan menggunakan lambda dan Reduce, temukan total waktu respons rata-rata untuk setiap neighborhood, waktu pengiriman rata-rata untuk setiap neighborhood, dan total waktu rata-rata untuk setiap neighborhood dan simpanlah ke dalam list dictionary. Tambahkan item dictioanary untuk menyertakan data populasi untuk semua Detroit dalam daftar gabungan Anda.


```python
data = list(filter(lambda x: x['zip_code'] and x['neighborhood'] and x['totalresponsetime'] and x['dispatchtime'] and x['totaltime'], data))
data_by_neighborhood = {}
for baris in data:
    if baris['neighborhood'] in data_by_neighborhood:
        data_by_neighborhood[baris['neighborhood']].append(baris)
    else:
        data_by_neighborhood[baris['neighborhood']] = [baris]

# Mengelompokkan data berdasarkan Neighborhood dan menghitung total waktu respons rata-rata, waktu pengiriman rata-rata, dan total waktu rata-rata untuk setiap neighborhood menggunakan lambda dan reduce
neighborhood_data = []
for neighborhood, data in data_by_neighborhood.items():
    total_waktu_resp_neighborhood = reduce(lambda acc, x: acc + float(x['totalresponsetime']), data, 0) / len(data)
    total_waktu_kirim_neighborhood = reduce(lambda acc, x: acc + float(x['dispatchtime']), data, 0) / len(data)
    total_waktu_neighborhood = reduce(lambda acc, x: acc + float(x['totaltime']), data, 0) / len(data)
    neighborhood_data.append({'Neighborhood': neighborhood, 'Rata-rata waktu respon': total_waktu_resp_neighborhood,
                              'Rata-rata waktu kirim': total_waktu_kirim_neighborhood, 'Total Time ': total_waktu_neighborhood})
```

```python
from tabulate import tabulate

# ...

# Mengubah list of dictionary menjadi list of list
table = [[d['Neighborhood'], d['Rata-rata waktu respon'], d['Rata-rata waktu kirim'], d['Total Time ']] for d in neighborhood_data]

# Mencetak tabel dengan menggunakan tabulate
print(tabulate(table, headers=['Neighborhood', 'Rata-rata waktu respon', 'Rata-rata waktu kirim', 'Total Time ']))
```
