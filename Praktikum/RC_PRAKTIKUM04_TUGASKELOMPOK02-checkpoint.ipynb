### Nomor 1
Bangun array sederhana dari fungsi yang mengambil input tunggal dan mengembalikan input yang ditambahkan dengan angka. Fungsi dibangun di loop for di mana iterator i berjalan dari 0 hingga 3. Untuk masing-masing angka ini dibuat fungsi lambda yang menangkap i dan menambahkannya ke input fungsi. Baris terakhir memanggil fungsi lambda kedua dengan 3 sebagai parameter. Yang hasilnya adalah 4.

```python
def input_tunggal(input):
    array_fungsi = []
    for i in range(4):
        array_fungsi.append(lambda x, i=i: x + i) # Fungsi lambda dibuat dengan variabel i sebagai input dan menambahkannya ke input fungsi
    return array_fungsi[1](input) # memanggil fungsi lambda kedua dengan 3 sebagai parameter

print(input_tunggal(3)) # memanggil fungsi lambda kedua dengan 3 sebagai parameter
```

```python
adders=[None, None, None, None]
for i in [0,1,2,3]:
    adders[i]=lambda a,i=i: i+a
print( adders[1](3) )
```

#### Menggunakan dataset breast cancer winconsin yang diimpor dari uci machine learning

```python
import pandas as pd

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data', header=None)

# Mengubah nilai target menjadi 1 untuk tumor ganas dan 0 untuk tumor jinak
df[1] = df[1].apply(lambda x: 1 if x=='M' else 0)

# Mengambil hanya kolom fitur numerik dan menghilangkan kolom target
df = df.iloc[:, 2:]

def counter(n, func, args):
    from datetime import datetime

    t1 = datetime.now()
    for i in range(n):
        func(*args)

    t2 = datetime.now()
    print("repeat {} times, cost time: {} seconds".format(n, (t2-t1).total_seconds()))

# Mengambil mean dari setiap kolom sebagai baseline
baseline = df.mean()

# Fungsi untuk menghitung jarak euclidean antara dua vektor
def euclidean_distance(a, b):
    return ((a-b)**2).sum()**0.5

# Double argument version
def func_1(row, baseline):
    return euclidean_distance(row, baseline)

# Partial function version
from functools import partial
func_2 = partial(func_1, baseline=baseline)

# Lambda function version
func_3 = lambda row: func_1(row, baseline)

# Closure version
def outer(baseline):
    def inner(row):
        return euclidean_distance(row, baseline)
    return inner

func_4 = outer(baseline)

# Functor version
class Functor():
    def __init__(self, baseline):
        self.baseline = baseline
        
    def __call__(self, row):
        return euclidean_distance(row, self.baseline)

func_5 = Functor(baseline)

# Currying version
def func_6(baseline):
    func = lambda row: func_1(row, baseline)
    return func

func_6 = func_6(baseline)

# Menghitung waktu eksekusi untuk masing-masing fungsi
counter(1000, func_1, [df.iloc[0], baseline])
counter(1000, func_2, [df.iloc[0]])
counter(1000, func_3, [df.iloc[0]])
counter(1000, func_4, [df.iloc[0]])
counter(1000, func_5, [df.iloc[0]])
counter(1000, func_6, [df.iloc[0]])
```

####  analisa closure

```python
def curry(f):
  # Menghitung jumlah argumen yang diterima oleh fungsi f
  argc=f._code_.co_argcount
  # Menyiapkan list untuk menampung argumen posisional
  f_args=[]
  # Menyiapkan dictionary untuk menampung argumen kata kunci
  f_kwargs={}
  # Membuat fungsi g sebagai decorator yang menerima argumen posisional dan kata kunci
  def g(*args,**kwargs):
    nonlocal f_args,f_kwargs
    # Menambahkan argumen posisional ke dalam list f_args
    f_args+=argc
    # Menambahkan argumen kata kunci ke dalam dictionary f_kwargs
    f_kwargs.update(kwargs)
    # Jika jumlah argumen posisional dan kata kunci yang diterima sudah sama dengan jumlah argumen yang dibutuhkan oleh fungsi f
    if len(f_args)+len(f_kwargs)==argc:
      # Panggil fungsi f dengan argumen yang sudah terkumpul
      return f(*f_args,**f_kwargs)
    else:
      # Jika belum, maka kembalikan fungsi g untuk menunggu argumen selanjutnya
      return g
  # Kembalikan fungsi g sebagai hasil dari dekorasi
  return g
```

#### composition function (aljabar)

```python
def compose(f, g):
    def h(*args, **kwargs):
        return f(g(*args, **kwargs))
    return h

def add_five(x):
    return x + 5

def square(x):
    return x * x

add_five_then_square = compose(square, add_five)

result = add_five_then_square(3)

def curry(f):
    argc = f.__code__.co_argcount  
    f_args = []
    f_kwargs = {}
    def g(*args, **kwargs):
        nonlocal f_args, f_kwargs
        f_args += args
        f_kwargs.update(kwargs)
        if len(f_args) + len(f_kwargs) >= argc:
            # Jika ya, maka memanggil fungsi f dengan argumen yang sudah dikumpulkan
            return f(*f_args, **f_kwargs)
        else:
            return g
    return g

print(result)
```

#### Analisis
Fungsi compose mengambil dua fungsi f dan g, dan mengembalikan fungsi h yang menerapkan fungsi f pada output fungsi g. Fungsi h menerima argumen apa pun, meneruskannya ke fungsi g, dan kemudian menerapkan fungsi f pada output dari fungsi g. Sebagai contoh, dengan menggunakan fungsi compose dan fungsi add_five dan square yang telah didefinisikan, kita bisa mendefinisikan add_five_then_square sebagai komposisi dari kedua fungsi tersebut, sehingga memanggil add_five_then_square(3) akan menghasilkan nilai (3 + 5) ** 2 = 64. <br>

compose dan curry adalah teknik dasar dalam pemrograman fungsional yang memungkinkan kita untuk menulis kode yang lebih ekspresif dan mudah dibaca. Dalam hal ini, compose memungkinkan kita untuk menggabungkan beberapa fungsi menjadi satu fungsi, sedangkan curry memungkinkan kita untuk memanggil fungsi dengan cara yang lebih fleksibel dan modular. Teknik-teknik ini dapat membantu kita menghindari pengulangan kode yang tidak perlu dan membuat kode kita lebih terstruktur dan mudah diubah.

#### Partial Application (Aljabar) 

```python
def add(a, b, c):
    return a + b + c

def partial_apply(f, *args):
    def wrapper(*more_args):
        return f(*args, *more_args)
    return wrapper

add_five = partial_apply(add, 5)

result = add_five(3, 2)


def curry(f):
    argc = f.__code__.co_argcount  
    f_args = []
    f_kwargs = {}
    def g(*args, **kwargs):
        nonlocal f_args, f_kwargs
        f_args += args
        f_kwargs.update(kwargs)
        if len(f_args) + len(f_kwargs) >= argc:
            # Jika ya, maka memanggil fungsi f dengan argumen yang sudah dikumpulkan
            return f(*f_args, **f_kwargs)
        else:
            return g
    return g

print(result)
```

#### Analisis 
Fungsi partial_apply mengambil sebuah fungsi f dan sejumlah argumen args, dan mengembalikan sebuah fungsi baru yang menerapkan fungsi f pada argumen args yang sudah ditentukan bersama dengan argumen tambahan apa pun yang diberikan pada saat pemanggilan fungsi baru. Dalam contoh di atas, kita menggunakan partial_apply untuk menghasilkan fungsi add_five, yang menerapkan fungsi add pada argumen 5 dan dua argumen tambahan lainnya yang diberikan pada saat pemanggilan. Sehingga, add_five(3, 2) akan menghasilkan nilai 10. <br>

partial_apply dan curry adalah teknik lain dalam pemrograman fungsional yang memungkinkan kita untuk menulis kode yang lebih modular dan mudah diubah. Dalam hal ini, partial_apply memungkinkan kita untuk membuat fungsi baru yang sudah memiliki beberapa argumen yang ditentukan terlebih dahulu, sehingga kita bisa menggunakan fungsi baru tersebut di banyak tempat tanpa harus menentukan argumen yang sama setiap saat. Sedangkan curry memungkinkan kita untuk memanggil fungsi dengan cara yang lebih modular dan fleksibel, seperti yang telah dijelaskan sebelumnya. Kedua teknik ini dapat membantu kita menghindari pengulangan kode yang tidak perlu dan membuat kode kita lebih terstruktur dan mudah dibaca.

 #### Teknik Currying

```python
def add(a):
    def add_b(b):
        return a + b
    return add_b

result = add(2)(3) # Output: 5

def curry(f):
    argc = f.__code__.co_argcount  
    f_args = []
    f_kwargs = {}
    def g(*args, **kwargs):
        nonlocal f_args, f_kwargs
        f_args += args
        f_kwargs.update(kwargs)
        if len(f_args) + len(f_kwargs) >= argc:
            # Jika ya, maka memanggil fungsi f dengan argumen yang sudah dikumpulkan
            return f(*f_args, **f_kwargs)
        else:
            return g
    return g

print(result)
```

#### Analisis
Fungsi add adalah contoh dari teknik penggunaan closure dalam Python. Fungsi ini mengambil satu argumen a dan mengembalikan fungsi baru add_b yang juga mengambil satu argumen b. Fungsi add_b akan mengembalikan hasil penjumlahan antara a dan b. Dalam contoh di atas, kita memanggil fungsi add dengan argumen 2, kemudian memanggil kembali fungsi yang dikembalikan add_b dengan argumen 3. Hasilnya adalah 5. <br>
