## 4.Apa yang dimaksud dengan pure functional programming language?

Pure functional programming language adalah bahasa pemrograman yang hanya menggunakan paradigma functional dan tidak memperkenankan adanya efek samping (side effect). Efek samping adalah suatu tindakan yang memengaruhi suatu state eksternal seperti mengubah nilai suatu variabel atau menulis ke dalam file. Dalam bahasa pemrograman functional pure, setiap fungsi harus bersifat deterministik, yaitu memiliki output yang sama untuk setiap input yang sama. Ini membuat program yang ditulis dalam bahasa pemrograman functional pure lebih mudah diprediksi dan dipahami. Contohnya bahasa pemrograman seperti Haskell, Scheme, dan Standard ML.

## 6. Membuat program Simple Password encryption, dengan ketentuan :
##### 1. Input password bertipe string dengan batasan panjang password 100 karakter
##### 2. Setiap karakter dari string tersebut diubah menjadi ASCII value
##### 3. Kemudian setiap ASCII value tersebut dikembangkan menjadi 3 value dengan dilakukan operasi berikut:
1. Value pertama dibagi dengan integer 26 kemudian dibulatkan dan dijumlahkan dengan 80 kemudian diubah kembali menjadi karakter <br>
2. Value kedua dihitung sisa bagi dengan 26 kemudian dijumlahkan dengan 80 kemudian diubah kembali menjadi  karakter <br>
3. Value ketiga adalah karakter ‘+’ jika value pertama lebih besar dari value kedua, atau value ketiga adalah karakter ‘-’ jika value pertama tidak lebih besar dari value kedua
##### 4. Deretkan value pertama, kedua dan ketiga untuk setiap karakter dalam password tersebut sehingga didapatkan password dalam bentuk baru

## Exercise 1

#### 3.a. Bantulah user tersebut dengan membuatkan sebuah program yang secara otomatis mengubah password yang di input menjadi terenkripsi sesuai aturan tersebut!
#### 3.b. Apa output yang dihasilkan dari program tersebut jika input password adalah ‘anakanakcerdas2020’ ?

```python
def encrypt_password(password):
    if len(password) > 100:
        return "Password too long. Maximum length is 100 characters."
    else:
        encrypted_password = []
        for char in password:
            ascii_value = ord(char)
            new_value_1 = (ascii_value // 26) + 80
            new_char_1 = chr(new_value_1)
            new_value_2 = (ascii_value % 26) + 80
            new_char_2 = chr(new_value_2)
            new_char_3 = '+' if new_value_1 > new_value_2 else '-'
            encrypted_password.extend([new_char_1, new_char_2, new_char_3])
        return ''.join(encrypted_password)
```

```python
password = input("Password: ")
encrypted_password = encrypt_password(password)
print("Encrypted password: ", encrypted_password)
```

#### 3.c. (Bonus) User tersebut lupa password asli yang dia inputkan ke dalam program tersebut, password setelah dienkripsi adalah ‘Sc-TV-Sc-TS+T[-Sc-TQ+TV-T[-Sf-Sc-T\-Sc-Qh-Qf-Qh-Qf-TS+Sg-Se-Sg-’. Bantulah user tersebut mendapatkan password asli nya!

```python
def decrypt_password(encrypted_password):
    password = ""
    i = 0
    while i < len(encrypted_password):
        val1 = ord(encrypted_password[i]) - 80
        val2 = ord(encrypted_password[i + 1]) - 80
        ascii_val = (val1 * 26) + val2
        password += chr(ascii_val)
        i += 3
    return password
```

```python
encrypted_password = input("encryption password:")
password = decrypt_password(encrypted_password)
print("Password: ", password)
```
